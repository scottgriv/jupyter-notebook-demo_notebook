```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 400)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y1, '-b', label='sin(x)')
plt.plot(x, y2, '-r', label='cos(x)')
plt.legend(loc='upper right')
plt.title('Plot of sin and cos from 0 to 10')
plt.grid()
plt.show()
```
